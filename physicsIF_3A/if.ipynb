```python
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import time

import os
import moviepy.video.io.ImageSequenceClip
image_folder='imgs/'
from PIL import Image, ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python

for i in range(0, 4):
    min=0
    max=10

    x1 = random.randint(min, max)
    y1 = random.randint(min, max)

    x2 = random.randint(min, max)
    y2 = random.randint(min, max)

    plt.scatter(x1,y1, c='red') # start 
    plt.scatter(x2,y2, c='green') # end

    dx = x2-x1
    dy = y2-y1

    plt.arrow(x=x1, y=y1, dx=dx, dy=dy, head_width = 0.05, color='black')
    
plt.xlabel("x")
plt.ylabel("y")    
plt.xlim(-0.5,11)
plt.ylim(-0.5,11)
plt.show()
```

```python
xnew = 0
ynew = 0
xstart = xnew
ystart = ynew

n_points = 23

figure, axes = plt.subplots( 1 )
circle = plt.Circle( (20, 20 ), 5 ,fill = False )
axes.set_aspect(1)
axes.add_artist(circle )
plt.axhline(y=20, color='black', linestyle='-', linewidth=0.5)
plt.xlim(-0.5,40)
plt.ylim(-0.5,40)
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(xnew, ynew, c='red', s=15)
plt.savefig('imgs/0.png')

for i in range(1, n_points+1):
    plt.scatter(xnew, ynew, c='red', s=15)
    
    dx = round(random.randint(0,3), 2)
    dy = round(random.randint(0,3), 2)

    xnew = xnew + dx
    ynew = ynew + dy
    
    plt.arrow(x=xstart, y=ystart, dx=dx, dy=dy, head_width = 0.08, color='black')
    
    xstart = xnew
    ystart = ynew
    
    print(f'x = {xnew}, y = {ynew}')

    plt.scatter(xnew, ynew, c='red', s=15)
    plt.xlim(-0.5,40)
    plt.ylim(-0.5,40)
    plt.xlabel("x [m]")
    plt.ylabel("y [m]")
    time.sleep(3) # Frame Per Second
    plt.axhline(y=20, color='black', linestyle='-', linewidth=0.5)
    plt.savefig(f'imgs/{i}.png')


plt.arrow(x=0, y=0, dx=xnew, dy=ynew, head_width = 0.1, color='blue')
plt.savefig(f'imgs/{n_points+1}.png')
plt.grid()
plt.show()
```

```python
image_files = []

for img_number in range(0,n_points+2): 
    image_files.append(image_folder + str(img_number) + '.png')
```

```python
fps = 5

clip = moviepy.video.io.ImageSequenceClip.ImageSequenceClip(image_files, fps=fps)
clip.write_videofile(image_folder + 'fullmovie.mp4')
```
